```
# =========================================================
# TALLER PRACTICO PANDAS + EXCEL
# ANYELA KATERINE RENTERIA CUAMA
# GOOGLE COLAB
# =========================================================

# Importamos la herramienta files de Google Colab
# para poder subir archivos Excel
from google.colab import files

# Subimos el archivo Excel al entorno de Colab
uploaded = files.upload()

# =========================================================
# IMPORTAR LIBRERIA
# =========================================================

# Importamos la librería pandas
# para trabajar con tablas y datos
import pandas as pd
```

```
# =========================================================
# LEER ARCHIVO EXCEL
# =========================================================

# Leemos el archivo Excel cargado
# y lo guardamos en el DataFrame df
df = pd.read_excel(next(iter(uploaded)))
```

```
# =========================================================
# MOSTRAR DATAFRAME
# =========================================================

# Imprimimos una línea decorativa
print("===================================")

# Mostramos el título
print("DATAFRAME COMPLETO")

# Imprimimos otra línea decorativa
print("===================================")

# Mostramos el DataFrame completo
display(df)
```

```
# =========================================================
# PRIMERAS 5 FILAS
# =========================================================

# Hacemos un salto de línea y mostramos título
print("\n===================================")

# Mostramos subtítulo
print("PRIMERAS 5 FILAS")

# Línea decorativa
print("===================================")

# Mostramos las primeras 5 filas
display(df.head())
```

```
# =========================================================
# ULTIMAS 10 FILAS
# =========================================================

# Mostramos línea decorativa
print("\n===================================")

# Mostramos subtítulo
print("ULTIMAS 10 FILAS")

# Mostramos línea decorativa
print("===================================")

# Mostramos las últimas 10 filas
display(df.tail(10))
```

```
# =========================================================
# TIPOS DE DATOS DE CADA COLUMNA
# =========================================================

# Mostramos línea decorativa
print("\n===================================")

# Mostramos subtítulo
print("TIPOS DE DATOS DE CADA COLUMNA")

# Mostramos línea decorativa
print("===================================")

# Mostramos el tipo de dato
# de cada columna del DataFrame
display(df.dtypes)
```

```
# =========================================================
# TAMAÑO DEL DATAFRAME
# =========================================================

# Mostramos línea decorativa
print("\n===================================")

# Mostramos subtítulo
print("TAMAÑO DEL DATAFRAME")

# Mostramos línea decorativa
print("===================================")

# Mostramos cantidad de filas
print("Cantidad de filas:")
print(df.shape[0])

# Mostramos cantidad de columnas
print("\nCantidad de columnas:")
print(df.shape[1])
```

```
# =========================================================
# DESCRIPCION ESTADISTICA
# =========================================================

# Mostramos línea decorativa
print("\n===================================")

# Mostramos subtítulo
print("DESCRIPCION ESTADISTICA")

# Mostramos línea decorativa
print("===================================")

# Mostramos estadísticas descriptivas
display(df.describe())
```

```
# =========================================================
# CREAR TOTAL_VENTA
# =========================================================

# Multiplicamos la columna Precio
# por la columna Cantidad
df['Total_Venta'] = df['Precio'] * df['Cantidad']

# Mostramos línea decorativa
print("\n===================================")

# Mostramos subtítulo
print("TOTAL_VENTA")

# Mostramos línea decorativa
print("===================================")

# Mostramos columnas seleccionadas
display(df[['Producto', 'Total_Venta']])
```

```
# =========================================================
# CREAR COSTO_TOTAL
# =========================================================

# Sumamos Total_Venta y Costo_Envio
df['Costo_Total'] = df['Total_Venta'] + df['Costo_Envio']

# Mostramos línea decorativa
print("\n===================================")

# Mostramos subtítulo
print("COSTO_TOTAL")

# Mostramos línea decorativa
print("===================================")

# Mostramos columnas seleccionadas
display(df[['Producto', 'Costo_Total']])
```

```
# =========================================================
# PROMEDIO
# =========================================================

# Calculamos promedio de Total_Venta
promedio = df['Total_Venta'].mean()

# Mostramos línea decorativa
print("\n===================================")

# Mostramos subtítulo
print("PROMEDIO TOTAL_VENTA")

# Mostramos línea decorativa
print("===================================")

# Mostramos promedio
print(promedio)
```

```
# =========================================================
# PRODUCTO MAYOR VENTA
# =========================================================

# Buscamos el producto
# con mayor Total_Venta
producto_mayor = df.loc[df['Total_Venta'].idxmax()]

# Mostramos línea decorativa
print("\n===================================")

# Mostramos subtítulo
print("PRODUCTO MAYOR VENTA")

# Mostramos línea decorativa
print("===================================")

# Mostramos resultado
display(producto_mayor)
```

```
# =========================================================
# PRODUCTO MENOR VENTA
# =========================================================

# Buscamos el producto
# con menor Total_Venta
producto_menor = df.loc[df['Total_Venta'].idxmin()]

# Mostramos línea decorativa
print("\n===================================")

# Mostramos subtítulo
print("PRODUCTO MENOR VENTA")

# Mostramos línea decorativa
print("===================================")

# Mostramos resultado
display(producto_menor)
```

```
# =========================================================
# CREAR COLUMNA IVA
# =========================================================

# Se crea una nueva columna llamada IVA
# El IVA corresponde al 19% del Total_Venta
df['IVA'] = df['Total_Venta'] * 0.19

# Mostrar resultados
print("\n===================================")

# Mostramos subtítulo
print("COLUMNA IVA")

# Línea decorativa
print("===================================")

# Mostramos Producto, Total_Venta e IVA
display(df[['Producto', 'Total_Venta', 'IVA']])
```

```
# =========================================================
# CREAR COLUMNA GANANCIA_ESTIMADA
# =========================================================

# Se calcula una ganancia estimada del 25%
# tomando como base el Total_Venta
df['Ganancia_Estimada'] = df['Total_Venta'] * 0.25

# Mostrar resultados
print("\n===================================")

# Mostramos subtítulo
print("GANANCIA ESTIMADA")

# Línea decorativa
print("===================================")

# Mostramos Producto y Ganancia_Estimada
display(df[['Producto', 'Ganancia_Estimada']])
```

```
# =========================================================
# CREAR COLUMNA PERDIDA_ESTIMADA
# =========================================================

# Se calcula una pérdida estimada del 5%
# basada en el Total_Venta
df['Perdida_Estimada'] = df['Total_Venta'] * 0.05

# Mostrar resultados
print("\n===================================")

# Mostramos subtítulo
print("PERDIDA ESTIMADA")

# Línea decorativa
print("===================================")

# Mostramos Producto y Perdida_Estimada
display(df[['Producto', 'Perdida_Estimada']])
```

```
# =========================================================
# PORCENTAJE DE VENTAS
# =========================================================

# Se calcula el total general de ventas
total_general = df['Total_Venta'].sum()

# Se calcula el porcentaje que representa
# cada producto respecto al total
df['Porcentaje_Ventas'] = (
    df['Total_Venta'] / total_general
) * 100

# Mostrar resultados
print("\n===================================")

# Mostramos subtítulo
print("PORCENTAJE DE VENTAS")

# Línea decorativa
print("===================================")

# Mostramos Producto y Porcentaje_Ventas
display(df[['Producto', 'Porcentaje_Ventas']])
```

```
# =========================================================
# PRODUCTOS CON PORCENTAJE MAYOR AL 15%
# =========================================================

# Se filtran únicamente los productos
# cuyo porcentaje de ventas es superior al 15%
productos_superiores = (
    df[df['Porcentaje_Ventas'] > 15]
)

# Mostrar resultados
print("\n===================================")

# Mostramos subtítulo
print("PRODUCTOS SUPERIORES AL 15%")

# Línea decorativa
print("===================================")

# Mostramos únicamente:
# Producto y Porcentaje_Ventas
print(
    productos_superiores[
        ['Producto', 'Porcentaje_Ventas']
    ]
)
```

```
# =========================================================
# MEDIANA DE PRECIO
# =========================================================

# La mediana representa el valor central
# de todos los precios ordenados

mediana_precio = df['Precio'].median()

# Mostrar resultados
print("\n===================================")

# Mostramos subtítulo
print("MEDIANA DE PRECIO")

# Línea decorativa
print("===================================")

# Mostramos la mediana
print(mediana_precio)
```

```
# =========================================================
# MEDIANA DE PRECIO
# =========================================================

# La mediana representa el valor central
# de todos los precios ordenados

mediana_precio = df['Precio'].median()

# Mostrar resultados
print("\n===================================")

# Mostramos subtítulo
print("MEDIANA DE PRECIO")

# Línea decorativa
print("===================================")

# Mostramos la mediana
print(mediana_precio)
```

```
# =========================================================
# MODA DE PRECIO
# =========================================================

# La moda representa el valor
# que más se repite

moda_precio = df['Precio'].mode()

# Mostrar resultados
print("\n===================================")

# Mostramos subtítulo
print("MODA DE PRECIO")

# Línea decorativa
print("===================================")

# Mostramos la moda
display(moda_precio)
```

```
# =========================================================
# DESVIACION ESTANDAR
# =========================================================

# La desviación estándar mide
# qué tan dispersos están los datos

desviacion = df['Cantidad'].std()

# Mostrar resultados
print("\n===================================")

# Mostramos subtítulo
print("DESVIACION ESTANDAR")

# Línea decorativa
print("===================================")

# Mostramos resultado
print(desviacion)
```

```
# =========================================================
# MAXIMO Y MINIMO
# =========================================================

# Valor máximo de ventas
maximo = df['Total_Venta'].max()

# Valor mínimo de ventas
minimo = df['Total_Venta'].min()

# Mostrar valor máximo
print("\n===================================")

# Mostramos subtítulo
print("VALOR MAXIMO")

# Línea decorativa
print("===================================")

# Mostramos valor máximo
print(maximo)

# Mostrar valor mínimo
print("\n===================================")

# Mostramos subtítulo
print("VALOR MINIMO")

# Línea decorativa
print("===================================")

# Mostramos valor mínimo
print(minimo)
```

```
# =========================================================
# EXPLICACION ESTADISTICA
# =========================================================

# Mostrar resultados
print("\n===================================")

# Mostramos subtítulo
print("EXPLICACION ESTADISTICA")

# Línea decorativa
print("===================================")

# Explicación de cada concepto
print("La media representa el promedio de los precios.")

print("La mediana representa el valor central.")

print("La moda representa el valor que más se repite.")

print("La desviación estándar mide la dispersión de los datos.")

print("El valor máximo representa la venta más alta.")

print("El valor mínimo representa la venta más baja.")
```

```
# =========================================================
# FILTRAR CATEGORIA TECNOLOGIA
# =========================================================

# Se filtran únicamente
# los productos de tecnología

tecnologia = df[df['Categoria'] == 'Tecnología']

# Mostrar resultados
print("\n===================================")

# Mostramos subtítulo
print("PRODUCTOS TECNOLOGIA")

# Línea decorativa
print("===================================")

# Mostramos productos filtrados
display(tecnologia)
```

```
# =========================================================
# ORDENAR PRODUCTOS
# =========================================================

# Se ordenan los productos
# desde la mayor venta hasta la menor

ordenados = df.sort_values(
    by='Total_Venta',
    ascending=False
)

# Mostrar resultados
print("\n===================================")

# Mostramos subtítulo
print("PRODUCTOS ORDENADOS")

# Línea decorativa
print("===================================")

# Mostramos productos ordenados
display(ordenados[['Producto', 'Total_Venta']])
```

```
# =========================================================
# EXPORTAR A EXCEL
# =========================================================

# Exportamos el DataFrame final
# a un archivo Excel
df.to_excel(
    'resultado_final_pandas.xlsx',
    index=False
)

# Mostramos línea decorativa
print("===================================")

# Mostramos subtítulo
print("ARCHIVO EXPORTADO")

# Mostramos línea decorativa
print("===================================")

# Mostramos mensaje de confirmación
print("Archivo Excel exportado correctamente")
```
