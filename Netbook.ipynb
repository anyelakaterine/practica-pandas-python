IMPORTACIÓN DE LIBRERÍAS
Explicación

En esta sección se importan las librerías necesarias para desarrollar el taller.

pandas: permite trabajar con tablas y datos.
google.colab.files: permite subir y descargar archivos en Google Colab.
Código

```
# =========================================================
# IMPORTAR LIBRERIAS
# =========================================================

# Importamos pandas para trabajar con tablas
import pandas as pd

# Importamos files para subir archivos en Colab
from google.colab import files
```

CARGA DEL ARCHIVO EXCEL
Explicación

En esta parte se carga el archivo Excel desde el computador hacia Google Colab.

La función files.upload() abre una ventana para seleccionar el archivo.

```
# =========================================================
# SUBIR ARCHIVO EXCEL
# =========================================================

# Subimos archivo Excel
uploaded = files.upload()
```

LEER EL ARCHIVO EXCEL
Explicación

La función pd.read_excel() permite leer el archivo Excel y convertirlo en un DataFrame.

Un DataFrame es una estructura de datos organizada en filas y columnas.

```
# =========================================================
# LEER ARCHIVO EXCEL
# =========================================================

# Leemos el archivo Excel cargado
df = pd.read_excel(next(iter(uploaded)))
```

MOSTRAR DATAFRAME
Explicación

El DataFrame contiene toda la información del archivo Excel organizada en formato de tabla.

Código

```
# =========================================================
# MOSTRAR DATAFRAME
# =========================================================

print("===================================")
print("DATAFRAME COMPLETO")
print("===================================")

# Mostramos DataFrame completo
print(df)
```

Interpretación de resultados

El resultado permite observar:

Todos los productos.
Los precios.
Cantidades.
Categorías.
Costos de envío.

Esto permite verificar que el archivo fue cargado correctamente.\

CREACIÓN DE TOTAL_VENTA
Explicación

Se crea una nueva columna multiplicando:

Precio × Cantidad.

```
# =========================================================
# CREAR TOTAL_VENTA
# =========================================================

df['Total_Venta'] = df['Precio'] * df['Cantidad']

print(df[['Producto', 'Total_Venta']])
```

Interpretación

Permite conocer el valor total vendido por cada producto.

CREACIÓN DE COSTO_TOTAL
Explicación

Se suma:

Total_Venta + Costo_Envio.
Código

```
# =========================================================
# CREAR COSTO_TOTAL
# =========================================================

df['Costo_Total'] = df['Total_Venta'] + df['Costo_Envio']

print(df[['Producto', 'Costo_Total']])
```

Interpretación

Permite calcular el costo final considerando el envío

ESTADÍSTICAS PRINCIPALES
Media

La media representa el promedio de los precios.

```
media_precio = df['Precio'].mean()

print(media_precio)
```

Mediana

La mediana representa el valor central de los datos.

mediana_precio = df['Precio'].median()

print(mediana_precio)

Moda

La moda representa el valor que más se repite.

```
moda_precio = df['Precio'].mode()

print(moda_precio)
```

Desviación estándar

Mide qué tan dispersos están los datos.

```
desviacion = df['Cantidad'].std()

print(desviacion)
```

FILTRAR PRODUCTOS TECNOLÓGICOS
Explicación

Se filtran únicamente los productos de tecnología.

Código

```
tecnologia = df[df['Categoria'] == 'Tecnología']

print(tecnologia)
```

Interpretación

Permite analizar únicamente productos pertenecientes a la categoría Tecnología.

ORDENAR PRODUCTOS
Explicación

Los productos se ordenan desde la mayor venta hasta la menor.

Código

```
ordenados = df.sort_values(
    by='Total_Venta',
    ascending=False
)

print(ordenados[['Producto', 'Total_Venta']])
```

Interpretación

Permite identificar cuáles productos generan mayores ingresos.

EXPORTAR RESULTADOS A EXCEL
Explicación

Código

```
# Exportamos archivo final
df.to_excel(
    'resultado_final_pandas.xlsx',
    index=False
)

print("Archivo exportado correctamente")
```
